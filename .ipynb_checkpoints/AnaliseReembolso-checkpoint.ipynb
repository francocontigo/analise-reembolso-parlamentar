### Top 5 parlamentares para serem investigados

Somando os valors de reembolso de cada senador na categoria: Locomoção, hospedagem, alimentação, combustíveis e lubrificantes.

```python
df_loc_investigacao = df_loc.groupby(["SENADOR"])["VALOR_REEMBOLSADO"].sum().sort_values(ascending=False)
df_loc_investigacao = pd.DataFrame(df_loc_investigacao, columns=["VALOR_REEMBOLSADO"])
df_loc_investigacao.head()
```

Identificando o total de informações faltantes informadas nos reembolsos

```python
df_loc_null = df_loc.loc[:, ["CNPJ_CPF", "FORNECEDOR", "DOCUMENTO", "DATA", 
                             "DETALHAMENTO"]].isnull().groupby(df_loc["SENADOR"]).sum()
df_loc_null["TOTAL_NULO"] = df_loc_null.sum(axis=1).sort_values(ascending=False)
df_loc_null.drop(["CNPJ_CPF", "FORNECEDOR", "DOCUMENTO", "DATA",
                 "DETALHAMENTO"], axis=1, inplace=True)
df_loc_null.sort_values("TOTAL_NULO", ascending=False, inplace=True)
df_loc_null.head()
```

Unificando o total do valor reembolsado e o total de ocorrências nulas em um único DataFrame.

```python
df_loc_investigacao_final = pd.merge(df_loc_investigacao, df_loc_null, on="SENADOR").sort_values(["VALOR_REEMBOLSADO", "TOTAL_NULO"], ascending=[False,False])
df_loc_investigacao_final.describe()
```

```python
df_loc_investigacao_final.head(30)
```

### Buscando outliers usando interquartil

```python
#Calculando quartis e interquartil
loc_nulo_q1 = np.percentile(df_loc_investigacao_final["TOTAL_NULO"], 25)
loc_nulo_q3 = np.percentile(df_loc_investigacao_final["TOTAL_NULO"], 75)
loc_nulo_iqr = loc_nulo_q3 - loc_nulo_q1
lower_loc_nulo = loc_nulo_q1 - 1.5*loc_nulo_iqr
upper_loc_nulo = loc_nulo_q1 + 1.5*loc_nulo_iqr

loc_valor_q1 = np.percentile(df_loc_investigacao_final["VALOR_REEMBOLSADO"], 25)
loc_valor_q3 = np.percentile(df_loc_investigacao_final["VALOR_REEMBOLSADO"], 75)
loc_valor_iqr = loc_valor_q3 - loc_valor_q1
lower_loc_valor = loc_valor_q1 - 1.5*loc_valor_iqr
upper_loc_valor = loc_valor_q3 + 1.5*loc_valor_iqr

expressao_loc_outliers = "(TOTAL_NULO >= @upper_loc_nulo) | (VALOR_REEMBOLSADO >= @upper_loc_valor)"


df_loc_investigacao_final_outliers = df_loc_investigacao_final.query(expressao_loc_outliers)

df_loc_investigacao_final_outliers
```

# Análise Parlamentar Especifico

```python
df_otto_alencar = df_cota_parlamentar[df_cota_parlamentar["SENADOR"].isin(["OTTO ALENCAR"])].sort_values("VALOR_REEMBOLSADO", ascending=False)
df_otto_alencar.describe()
```

```python
agg_fun ={
    'VALOR_REEMBOLSADO': 'sum'
}
df_otto_alencar_ano = df_otto_alencar.groupby('ANO', as_index=False).agg(agg_fun)
sns.lineplot(x='ANO', y='VALOR_REEMBOLSADO', data = df_otto_alencar_ano).set(title="Reembolsos por semestre do Senador Otto Alencar,\n entre 2015 a 2019",
                                                                            ylabel="Valor reembolsado", xlabel="Semestres")
```

```python
# Gráfico do total de reembolsos do senador por categoria, não possui gastos com segurança
df_otto_alencar_despesas_por_tipo = df_otto_alencar.groupby(["TIPO_DESPESA"])["VALOR_REEMBOLSADO"].sum().sort_values(ascending=True)
df_otto_alencar_despesas_por_tipo.plot(kind="bar", title="Valor reembolsado por tipo de despesa do Senador Otto Alencar, \n entre 2015 a 2019, segundo dados públicos")
plt.xlabel("Tipo de despesas")
plt.ylabel("Valor reembolsado")
plt.xticks([0, 1, 2, 3, 4, 5], ["Escritório",
                                  "Divulgação",
                                  "Consultorias",
                                  "Passagens",
                                  "Imóveis",
                                  "Locomoção"],
          rotation=20)

plt.show()
```

Distribuição de reembolsos

```python
# Distribuição 
sns.histplot(df_otto_alencar["VALOR_REEMBOLSADO"], kde=True).set(xlabel="Valor Reembolsado", ylabel="Quantidade de Reembolsos", title="Distribuição de reembolsos por ticket do Senador Otto Alencar,\n entre 2015 a 2019, segundo dados públicos")
```

```python
df_otto_alencar_outliers = df_otto_alencar.nlargest(100, "VALOR_REEMBOLSADO").sort_values("VALOR_REEMBOLSADO", ascending=False)
df_otto_alencar_outliers
```

```python
# Gráfico de quantidade de dados faltantes do parlamentar.
df_otto_alencar[["TIPO_DESPESA", "CNPJ_CPF", "FORNECEDOR", "DOCUMENTO", "DATA", "DETALHAMENTO"]].isna().sum().plot(kind="bar")
plt.xlabel("Tipo de dado")
plt.ylabel("Quantidade de faltantes")
plt.xticks([0, 1, 2, 3, 4, 5], ["Categoria",
                                           "CPF/CNPJ",
                                           "Fornecedor",
                                           "Documento",
                                           "Data",
                                           "Detalhamento",],
          rotation=30)

plt.show()
```

Em comparação aos demais senadores, Otto Alencar apresenta um ticket médio de reembolso abaixo da média. Os maiores valores de reembolso são registrados na seguinte categoria: locomoção, hospedagem, alimentação, combustíveis e lubrificantes. Contudo, há uma quantidade significativa de dados faltantes na coluna detalhamento.
